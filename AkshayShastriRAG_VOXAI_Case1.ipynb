# Model Download

```python
# Import transformer classes for generaiton
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer
# Import torch for datatype attributes
import torch
```

```python
# Define variable to hold llama2 weights naming
name = "meta-llama/Llama-2-7b-chat-hf"
# Set auth token variable from hugging face
auth_token = XXX
```

```python
# Create tokenizer
tokenizer = AutoTokenizer.from_pretrained(name,
    cache_dir='./model/', use_auth_token=auth_token)
```

```python
# Create model
model = AutoModelForCausalLM.from_pretrained(name,
    cache_dir='./model/', use_auth_token=auth_token, torch_dtype=torch.float16,
    rope_scaling={"type": "dynamic", "factor": 2}, load_in_8bit=True)
```

# RAG Implementation

```python
# Import the prompt wrapper but for llama index
from llama_index.legacy.prompts.prompts import SimpleInputPrompt
# Create a system prompt
system_prompt = """<s>[INST] <<SYS>>
Your goal is to provide answers acting as the assitant taking orders for KFC in a drive thru.
This is vey important - If you don't know the answer to a question, please don't share false information.
This is very important - if an item is not present on the menu please dont add false information and say that it is present. Also dont add any information not explicitly mentioned in the context in your answer.
If you need more information in case of consfusion, ask for it from the customer.
Dont assume anything about the size of the items if it is not mentioned
Always mention the availability of the item which is being enquired about. if something is not availble suggest similar items from the menu only.
In the json provided for context, the price is mentioned on the next line after the name of the menu item.

Answer as the person taking the order only based on the context. Do not add outside information, and mention the availability, price and the nutrient information<</SYS>>
"""
# Throw together the query wrapper
query_wrapper_prompt = SimpleInputPrompt("{query_str} [/INST]")
```

```python
# Complete the query prompt
query_wrapper_prompt.format(query_str='hello')
```

```python
# Import the llama index HF Wrapper
from llama_index.legacy.llms import HuggingFaceLLM
# Create a HF LLM using the llama index wrapper
llm = HuggingFaceLLM(context_window=4096,
                    max_new_tokens=256,
                    system_prompt=system_prompt,
                    query_wrapper_prompt=query_wrapper_prompt,
                    model=model,
                    tokenizer=tokenizer,)
```

```python
# Bring in embeddings wrapper
from llama_index.legacy.embeddings import LangchainEmbedding
# Bring in HF embeddings - need these to represent document chunks
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
```

```python
# Create and dl embeddings instance
embeddings=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
)
```

```python
# Bring in stuff to change service context
from llama_index.legacy import set_global_service_context
from llama_index.legacy import ServiceContext
```

```python
# Create new service context instance
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embeddings
)
# And set the service context
set_global_service_context(service_context)
```

```python
# Import deps to load json
from llama_index.legacy import VectorStoreIndex, download_loader
from pathlib import Path
```

```python
JSONReader = download_loader("JSONReader")

loader = JSONReader()
documents = loader.load_data(Path('/content/menu.json'))
```

```python
# Create an index - updateable as per the conditions of the case
index = VectorStoreIndex.from_documents(documents)
```

```python
# Setup index query engine using LLM
query_engine = index.as_query_engine()
```
